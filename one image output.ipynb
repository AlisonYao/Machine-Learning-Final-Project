```
import tensorflow as tf
```

```
# read one pic
img = tf.io.read_file('000_1_0001.png')
# decode to tensor datatype with three channels
one_image = tf.image.decode_png(img, channels=3) # this is a tensor
```

```
# img = tf.image.resize(img, [60, 60])
```

```
saturated = tf.image.adjust_saturation(one_image, 3)
```

```
# change datatype
output_image = tf.cast(saturated, dtype=tf.uint8)
```

```
# encode to png
out = tf.image.encode_png(output_image)
```

```
# save
with tf.io.gfile.GFile('out.png', 'wb') as file:
    file.write(out.numpy())
```
